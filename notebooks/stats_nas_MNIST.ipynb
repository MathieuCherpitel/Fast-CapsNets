```python

import numpy as np
import json


acc = []
inf = []


for file in file_list:
    with open(f'../saved_ga/{file}/solutions.json') as json_file:
        data = json.load(json_file)

        for sol in data.items():
            acc.append(sol[1]['accuracy'])
            inf.append(round(sol[1]['inference'], 3))

acc = np.array(acc)
inf = np.array(inf)

acc, inf
```

```python
pd.DataFrame({'accuracy': acc, "inference": inf}).describe().round(3)
```
